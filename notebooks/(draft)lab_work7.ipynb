Примеры как можно задавать уравнения

Мини инструкция:
1) Сначала типа объявляем переменнуые
2) потом создаем класс самой задачи
3) и пихаем её в солвер

+ всё супер понятно и user-friendly
+ можно задать любую систему какую хотим, не сильно утруждаясь

- много кода под котом котооырй это парсит
- все равно на уровне солвера, будем работать с массивом  [ya, ya_'] и [yb, yb_']

```python
import numpy as np

from comp_math.bvp.model.bvp_problem import BVProblem
from comp_math.bvp.model.model_bvp import Derivative, Variable


print("Пример 1: y'' = -y, y(0)=1, y(1)=0")
y = Variable("y")

problem = BVProblem(
    equations=[Derivative(y, 2) == -y],
    conditions=[y.at(0) == 1, y.at(1) == 0],
    domain=(0, 1)
)

sol = problem.solve()
sol.plot(y)
sol.print_summary()

# Пример 2: С пользовательской функцией
print("\nПример 2: y'' = -sin(y) + 0.1*x*y', y(0)=1, y(1)=0")

def my_force(x, y, yp):
    return -np.sin(y) + 0.1 * x * yp

y = Variable("y")
y2 = Variable("y")

problem2 = BVProblem(
    equations=[Derivative(y2, 2) == my_force],
    conditions=[y2.at(0) == 1, y2.at(1) == 0],
    domain=(0, 1)
)

sol2 = problem2.solve()
sol2.plot(y2)

# Пример 3: Условия Робена
print("\nПример 3: y'' = -y, y'(0) + 2y(0) = 3, y(1) = 1")

y = Variable("y")
y2 = Variable("y")
y3 = Variable("y")

problem3 = BVProblem(
    equations=[Derivative(y3, 2) == -y3],
    conditions=[
        Derivative(y3, 1).at(0) + 2 * y3.at(0) == 3,
        y3.at(1) == 1
    ],
    domain=(0, 1)
)

sol3 = problem3.solve()
sol3.plot(y3)
sol3.plot(Derivative(y3, 1))

# Пример 4: Система уравнений
print("\nПример 4: u' = v, v' = -u, u(0)=1, v(1)=0")

u = Variable("u")
v = Variable("v")

problem4 = BVProblem(
    equations=[
        Derivative(u, 1) == v,
        Derivative(v, 1) == -u
    ],
    conditions=[
        u.at(0) == 1,
        v.at(1) == 0
    ],
    domain=(0, 1)
)

sol4 = problem4.solve()
sol4.plot(u)
sol4.plot(v)
```
